## Game functions

- Check position

- Check player won

```python
def display():
    print(li[7], '|',li[8], "|",li[9])
    print("-"*14)
    print(li[4], "|",li[5],"|",li[6])
    print("-"*14)
    print(li[1],"|",li[2],"|",li[3])
```

```python
def empty_position():
    ans=[]
    for i in range(1,len(li)):
        if li[i]==" ":
            ans.append(i)
    return ans
```

```python
def put_marker(position,marker):
    li[position]=marker
```

```python
def check_win():
    if (li[7]==li[8]==li[9] !=" ") or\
    (li[4]==li[5]==li[6] !=" ") or\
    (li[1]==li[2]==li[3] !=" ") or\
    (li[7]==li[4]==li[1] !=" ") or\
    (li[8]==li[5]==li[2] !=" ") or\
    (li[9]==li[6]==li[3] !=" ") or\
    (li[7]==li[5]==li[3] !=" ") or\
    (li[9]==li[5]==li[1] !=" "):
        return True
    else:
        return False
```

```python
def board_full():
    if len(empty_position())==0:
        return True
    else:
        return False
```

```python
def change_marker(marker):
    if marker=="0":
        return "X"
    else:
        return "0"
```

```python
def take_position():
    while True:
        user_input= input("Enter position(1-9) or 'quit' to exit :")
        if user_input.lower() in ("quit","exit"):
            return None
        try:
            position=int(user_input)
            if 1<=position<=9 and li[position] ==" ":
                return position
            else:
                print("position taken or invalid. Try again")
        except ValueError:
            print("please enter a number between 1-9 or 'quit' to exit.")
```

```python
from IPython.display import clear_output
import time
```

```python
li=[" "]*10
total_rounds=3
game=True
round_number=1
player1 , player2="X","O"
current_player=player1
wins_player1=0
wins_player2=0
time_player1=0
time_player2=0
game_start_time=time.time()
past_boards=[]
display()
while round_number <=total_rounds:
    li=[" "]*10
    current_player=player1
    round_start_time=None
    round_game=True
    
    while round_game:
         print (f"\n player {1 if current_player == player1 else 2}'s turn ({current_player})'")

         if round_start_time is None:
             
             round_start_time=time.time()
             
         turn_start_time = time.time()
         position=take_position()
         if position is None:
            total_game_time=time.time()-game_start_time
            display()
            print("game quit by player!")
            print(f"time taken by player 1 (X) : {time_player1: .2f} seconds" )
            print(f"time taken by player 2 (O) : {time_player2: .2f} seconds" )
            print(f"total game time : {total_game_time: .2f} seconds")
            round_game=False
            break
            
         put_marker(position,current_player)
         turn_end_time = time.time()
         turn_time = turn_end_time - turn_start_time
            
         if current_player== player1:
             time_player1 += turn_time
         else:
             time_player2+= turn_time 
         display()
         clear_output()
         if check_win():
             total_round_time=time.time()-round_start_time
             past_boards.append(li.copy())
             print(f"====congratulations {current_player} won the game==")
             if current_player==player1:
                 wins_player1+=1
             else:
                 wins_player2+=1
            
             print(f"scores - player 1 (X): {wins_player1}, player 2 (O) :{wins_player2}")
             print(f"time taken by player 1 (X) in round{round_number} : {time_player1: .2f} seconds" )
             print(f"time taken by player 2 (O)  in round{round_number} : {time_player2: .2f} seconds" )
             print(f"total time for round {round_number} : {total_round_time: .2f} seconds")
             round_game=False
             break  
            
         elif board_full():
             total_round_time=time.time()-round_start_time
             past_boards.append(li.copy())
             display()
             print(f"scores - player 1 (X): {wins_player1} , player 2 (O) :{wins_player2}")
             print(f"time taken by player 1 (X) in round{round_number} : {time_player1: .2f} seconds" )
             print(f"time taken by player 2 (O)  in round{round_number} : {time_player2: .2f} seconds" )
             print(f"total time for round {round_number} : {total_round_time: .2f} seconds")
             round_game=False
             break  
         else:
             current_player=player2 if current_player==player1 else player1
             display()
    round_number+=1
    
# final result after all rounds
if round_number>total_rounds:
    total_game_time=time.time()-round_start_time
    print("\n===Final Result===")
    print(f"\n player 1 (X) wins: {wins_player1} rounds")
    print(f"\n player 2 (O) wins: {wins_player2} rounds")
    print(f"\n time taken by player 1 (X) in round{round_number} : {time_player1: .2f} seconds" )
    print(f"\n time taken by player 2 (O)  in round{round_number} : {time_player2: .2f} seconds" )
    print(f"\n total game time: {total_game_time: .2f} seconds")
    if wins_player1 > wins_player2:
        print("player 1 (X) is the overall winner!")
    elif wins_player2 > wins_player1:
        print("player 2 (O) is the overall winner!")
    else:
        print("the game is draw overall")
        
```
